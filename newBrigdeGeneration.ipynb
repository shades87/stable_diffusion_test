This file is to generate new images of concrete bridges with various degrees of damage.
Damage that we are trying to generate are cracks and spalling of the concrete.
The images should be realistic photos, and at a distance that would be useful to train a semantic segmentation model that detects damage in concrete buildings.

```python
#import stable diffusion and makes sure that we are creating images on out GPU
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"
device = "cuda"
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
#here we should generate prompts with an LLM
```

```python
for i in range(10):
    prompt = "Realistic photo of a concrete bridge, the bridge is close to the photo"
    image = pipe(prompt).images[0]  
        
    image.save("bridge"+str(i)+".png")

    prompt = "Realistic photo of a concrete bridge, the bridge is close to the photo. The bridge has small amounts of cracks."
    image = pipe(prompt).images[0]  
        
    image.save("bridge_with_cracks"+str(i)+".png")

    prompt = "Realistic photo of a concrete bridge, the bridge is close to the photo. The bridge has small amounts of cracks and spalling."
    image = pipe(prompt).images[0]  
        
    image.save("structure_with_cracks_and_spalling"+str(i)+".png")

    prompt = "Realistic photo of a concrete bridge, the bridge is close to the photo. The bridge has significant amounts of cracks and spalling."
    image = pipe(prompt).images[0]  
        
    image.save("structure_with_significant_cracks_and_spalling"+str(i)+".png")
```
